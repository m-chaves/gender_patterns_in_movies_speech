```python
# Look at the closest words in the latent 
import gensim 
w2v = gensim.models.KeyedVectors.load("results/cornell_prepro.wordvectors")
most_similar = w2v.most_similar(positive=['women'])
print('women', [word for word, sim in most_similar])
most_similar = w2v.most_similar(positive=['men'])
print('men', [word for word, sim in most_similar])
most_similar = w2v.most_similar(positive=['robot'])
print('robot', [word for word, sim in most_similar])
most_similar = w2v.most_similar(positive=['cat'])
print('cat', [word for word, sim in most_similar])
most_similar = w2v.most_similar(positive=['dog'])
print('dog', [word for word, sim in most_similar])
most_similar = w2v.most_similar(positive=['food'])
print('food', [word for word, sim in most_similar])
```

```python
# Look at the most similar words
most_similar = w2v.most_similar(positive=['woman'])
print('woman', [word for word, sim in most_similar])
most_similar = w2v.most_similar(positive=['man'])
print('man', [word for word, sim in most_similar])
```
